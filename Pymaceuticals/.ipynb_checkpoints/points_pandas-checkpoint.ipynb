```python
#Magic function notebook
%matplotlib notebook

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata = "data/Mouse_metadata.csv"
study_results = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)

# Combine the data into a single dataset
all_data = pd.merge(mouse_metadata, study_results, how='outer', on='Mouse ID')
all_data.head()
```

```python
#Group data by drug regimen
#drug_groups = all_data.groupby('Drug Regimen')
#Get a count of data points per drug regimen
name = all_data['Drug Regimen'].unique()
count = all_data['Drug Regimen'].value_counts()
```

```python
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd',
          '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
#create the bar chart using pandas
count.plot(kind='bar', figsize=(7,5), color=colors, title='Data Points per Drug Regimen')
```

```python
plt.tight_layout()
plt.show()
```
