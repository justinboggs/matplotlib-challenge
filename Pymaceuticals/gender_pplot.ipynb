```python
#Magic function notebook
%matplotlib notebook

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata = "data/Mouse_metadata.csv"
study_results = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)

# Combine the data into a single dataset
all_data = pd.merge(mouse_metadata, study_results, how='outer', on='Mouse ID')
all_data.head()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pyplot
sex = all_data['Sex'].unique()
count = all_data['Sex'].value_counts()
count
```

```python
plt.pie(count, labels=sex, autopct=lambda p: '{:.2f}%({:.0f})', startangle=140)
```
